```python
from pycaret.classification import *
import pandas as pd
```

```python
data = pd.read_csv('loan_data.csv')
```

```python
data.head()
```

```python
clf = setup(data=data, target='loan_status', session_id=123)
```

```python
best_model = compare_models()
```

```python
final_model = finalize_model(best_model)
```

```python
save_model(final_model, 'pycaret_best_model')
```

```python
# Plot the performance of the best model
plot_model(best_model, plot='auc')      # AUC-ROC Curve
plot_model(best_model, plot='confusion_matrix')  # Confusion Matrix
plot_model(best_model, plot='feature')  # Feature Importance
plot_model(best_model, plot='learning')  # Learning Curve
```

```python
interpret_model(best_model)
```
