## 🌿 4. We now split into train and test 

```python
#Split the data into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X_scalled,Y_scalled,test_size=0.2,
                                                    random_state = 23)
```

```python
Y_train.shape
```

## ⚙️ 5. Now we fit the models and find the most appropriate

```python
#Fit
LR_model = LinearRegression()

LR_model.fit(X_train,Y_train)
```

```python
LR_predictions_test = LR_model.predict(X_test)
```

```python
LR_mse = mean_squared_error(Y_test,LR_predictions_test)
LR_mse
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, LR_predictions_test, alpha=0.6, color='blue', label='Predictions')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--', label='Perfect Fit')  # Ideal line
plt.xlabel('Actual Values (Y_test)')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```

```python
# Flatten the arrays
Y_test_flat = Y_test.ravel()  # Convert (5908, 1) to (5908,)
predictions_test_flat = LR_predictions_test.ravel()

# Create the DataFrame
results = pd.DataFrame({
    'Actual Values': Y_test_flat,
    'Predicted Values': predictions_test_flat
})

# Display the first few rows
print(results.head(10))
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, RandomForest_model_predictions, alpha=0.6, color='blue', label='Predictions')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--', label='Perfect Fit')  # Ideal line
plt.xlabel('Actual Values (Y_test)')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```

## Elastic net

```python
Elastic_net_model = ElasticNet(alpha=0.1, l1_ratio=0.1, random_state=42)
Elastic_net_model.fit(X_train,Y_train)                               
```

```python
Elastic_net_predict = Elastic_net_model.predict(X_test)
```

```python
param_grid = {
    'alpha': [0.1, 0.5, 1.0, 5.0, 10.0],
    'l1_ratio': [0.1, 0.5, 0.7, 0.9, 1.0]
}

grid_search = GridSearchCV(ElasticNet(random_state=42), param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, Y_train)

print(f"Best parameters: {grid_search.best_params_}")
```
