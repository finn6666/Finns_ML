```python
X = data[['longitude','latitude',
          'housing_median_age','total_rooms','total_bedrooms',
          'population','households','median_income','ocean_proximity']]
y = data[['median_house_value']]
```

```python


# Assuming you already have X, y, w, b, learning_rate, and epochs defined
X = np.random.rand(100, 9)  # Example: 100 samples, 9 features
y = np.random.rand(100)  # Example: 100 target values

w = np.zeros(X.shape[1])  # Weight vector of size 9 (one for each feature)
b = 0  # Bias term
learning_rate = 0.01
epochs = 100

# List to store the loss values
losses = []

def compute_loss(X, y, w, b):
    predictions = np.dot(X, w) + b
    loss = np.mean((predictions - y) ** 2)
    return loss

for epoch in range(epochs):
    for i in range(len(X)):
        xi = X[i:i+1]  # xi should be shape (1, 9) - One row, all 9 features
        yi = y[i:i+1]  # yi should be shape (1,) - One target value
        
        # Prediction and error for a single sample
        prediction = np.dot(xi, w) + b
        error = prediction - yi
        
        # Gradients
        gradient_w = 2 * xi.T.dot(error)  # Shape: (9, 1)
        gradient_b = 2 * error  # Scalar
        
        # Update weights and bias
        w -= learning_rate * gradient_w.flatten()  # Flatten gradient_w to match shape of w
        b -= learning_rate * gradient_b

    # Compute the loss for the current epoch and store it
    loss = compute_loss(X, y, w, b)
    losses.append(loss)
    
    # Print loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss}")

# Plotting the loss
plt.plot(range(epochs), losses, label="Loss over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Stochastic Gradient Descent Loss")
plt.grid(True)
plt.show()
```
